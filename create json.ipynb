```python
import torch
import matplotlib.pyplot as plt
import numpy
from tqdm.auto import tqdm
```

```python
import glob
```

```python
x = glob.glob("Annotations/*")
```

```python
x[0].split("\\")
```

```python
x2 = {}
for i in x:
    lab = i.split("\\")[-1]
    v = glob.glob(f"{i}/*")
    x2[lab] = v
```

```python
x2
```

```python
import scipy
```

```python
d = scipy.io.loadmat("Annotations/accordion/annotation_0010.mat")
```

```python
d
```

```python
processed = {}
for label,files in x2.items():
    v2 = []
    for i in files:
        name = i.split("\\")[-1]
        data = scipy.io.loadmat(i)
        coords = data["box_coord"]
        v2.append({
            "img_name": name.replace("annotation","image").replace("mat","jpg"),
            "coords": coords[0].tolist()
        })
    processed[label] = v2
```

```python
processed
```

```python
import json
```

```python
with open("processed_files.json","w") as file:
    json.dump(processed,file)
```
